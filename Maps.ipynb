```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Denmark = gpd.read_file("KOM_MULTIPART.shp")
```

```python
# from: https://www.energidataservice.dk/dataset/communityproduction/resource_extract/1ed52ea0-6f73-4067-a89e-43dd41fee45b?page=1&applied-filters=Month%7C%3D%7C2018-09-01+00%3A00#resource-preview
power = pd.read_excel("Production_per_Municipality_Data-23ea948e940b4fe28ded776473088717.xlsx", dtype = {"MunicipalityNo": str})
power["KOMKODE"] = str(0)+power["MunicipalityNo"]
merge = Denmark.merge(power, how = "left", on = "KOMKODE")
```

```python
power.Month.min()
```

```python
merge.head()
merge = merge.fillna(0)
merge["windpower"] = merge["OnshoreWindPower"]+merge["OffshoreWindPower"]
```

```python
merge.head()
```

```python
fig, ax = plt.subplots(figsize = (20,10))
merge.plot(ax = ax, edgecolor = "black", column = "windpower",  cmap = "Purples")

vmin = min(merge.windpower)
vmax = max(merge.windpower)
# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap="Purples", norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
plt.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (20,10))
merge.plot(ax = ax, edgecolor = "black", column = "windpower",  scheme = "quantiles", cmap = "Purples", legend = True)
plt.axis("off")
#plt.savefig(fname = "DanmarksKort.png", bbox_inches = "tight")
plt.show()
```
